Congratulations, you've got a job at Python Pizza. Your first job is to build an automatic pizza order program. Based on a user's order, work out their final bill.
 ``` Small Pizza: $15 ``` ``` 
``` Medium Pizza: $20 ``` 
``` Large Pizza: $25 ``` 
If they want to add pepperoni :
``` Pepperoni for Small Pizza: +$2 ``` 
``` Pepperoni for Medium or Large Pizza: +$3 ```
If they want cheese :
 ``` Extra cheese for any size pizza: + $1 ```
Instructions
1.	Open a new file in your Python development environment or IDE.
2.	Begin by typing the line 'Welcome to Python Pizza Deliveries' which will display a welcome message when the code is run.
3.	Add a line that prompts the user for the size of the pizza they want using the input() function and stores the user's input in a variable named 'size'.
4.	Add a line that prompts the user if they want pepperoni on their pizza using the input() function and stores the user's input in a variable named 'add_pepperoni'.
5.	Add a line that prompts the user if they want extra cheese on their pizza using the input() function and stores the user's input in a variable named 'extra_cheese'.
6.	Check your code for any syntax errors or mistakes.
7.	Run the code and test it by providing inputs for size, pepperoni, and extra cheese.
Example Input
size = "L"
add_pepperoni = "Y"
extra_cheese = "N"
Example Output
Your final bill is: $28.
When you hit **run**, this is what should happen



````python

print('Welcome To Python Pizza Deliveries')

print('Dear Esteemed Customer, our Pizzas come in three sizes:\n1. Small\n2. Medium\n3. Large\n',
      'Pizza prices are: \nS = ₦15, M = ₦20, or L = ₦25\nPepperoni topping for S = ₦2, M = ₦3, L = ₦3\nExtra Cheese will incur an additional cost of ₦1')

# Pizza prices
pizza_prices = {
    'S': 15,
    'M': 20,
    'L': 25
}

# Pepperoni Topping Prices
topping_prices = {
    'S': 2,
    'M': 3,
    'L': 3
}

# Price of extra cheese
extra_cheese_price = 1

# Loop to take order
while True:
    size = input("What size of pizza do you prefer? S, M, or L: ").upper()
    if size in pizza_prices:
        break
    else:
        print("We do not have this size. Please select S, M, or L")

# Ask for quantity of pizzas
while True:
    more_pizza = input('What quantity of pizza would you like? Select 1, 2, 3, 4, or 5: ')
    if more_pizza in ['1','2', '3', '4', '5']:
        print("Your order is on the way") 
        break
       
    else:
        print('please select options 1 - 5')
        break
# Ask user if they want pepperoni
while True:
    add_pepperoni = input("Would you like us to add pepperoni to your pizza? Y or N: ").upper()
    if add_pepperoni == 'Y' or add_pepperoni == 'N':
        break
    else:
        print("Please enter Y or N")

# Ask user if they want extra cheese
while True:
    add_extra_cheese = input("Do you want extra cheese? Y or N: ").upper()
    if add_extra_cheese == 'Y' or add_extra_cheese == 'N':
        print('Perfect! Your order will soon be ready')
        break
else:
        print("Please enter Y or N")

# Calculate the bill based on user's inputs
bill = pizza_prices[size] * int(more_pizza)
if add_pepperoni == 'Y':
    bill += topping_prices[size] * int(more_pizza)
if add_extra_cheese == 'Y':
    bill += extra_cheese_price * int(more_pizza)

print(f"Your final bill is: ₦{bill}.")
print("You can make payments using our payment channels. We also accept MasterCard and Visa cards.")
print("Thank you for your patronage!")
````
